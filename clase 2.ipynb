```python
# Para un circulo
th = np.linspace(0,2*np.pi)
x , y = np.cos(th) , np.sin(th)
plt.plot(x,y)
plt.gca().set_aspect('equal')
plt.show()
```

```python
#para una espiral
th = np.linspace(0,2*np.pi)
x , y = np.cos(th) , np.sin(th)
x_esp , y_esp = th/(2*np.pi)*np.cos(th) , th/(2*np.pi)*np.sin(th)
plt.plot(x,y,label=r'Circulo')
plt.plot(x_esp,y_esp,label=r'Espiral')
plt.gca().set_aspect('equal')
plt.show()
```

```python
def E_q(r,r_q,q):
    d32 = ((r[0]-r_q[0])**2 + (r[1]-r_q[1])**2 )**1.5
    return (q*(r[0]-r_q[0]))/d32 , (q*(r[1]-r_q[1]))/d32
```

```python
# linea 
def E_tot(r,r_qs,qs):
    Ex ,Ey = np.zeros_like(r[0]) , np.zeros_like(r[1])
    for i in range(len(qs)):
        E = E_q(r,r_qs[i],qs[i]) #Me regresa 2 variables
        Ex = Ex + E[0]
        Ey = Ey + E[1]
    return Ex,Ey
```
